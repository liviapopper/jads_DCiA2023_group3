```python
import pandas as pd

df = pd.read_csv('paragraphs_dataset_with_actors.csv')
df.head(50)
```

```python
def filter_min_number_actors():
    df["actors_list"] = df["actors_in_paragraph"].apply(lambda x: x.strip('[]').replace("'", "").split(', '))
    df['2_plus_actors'] = df["actors_list"].apply(lambda x: check_if_list_has_enough_nonempty_items(x))
    
def check_if_list_has_enough_nonempty_items(lst):
    if len(list(filter(None, lst))) >= 2:
        return True
    else:
        return False
```
